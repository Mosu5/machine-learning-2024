```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Salary%20Data.csv')
```

```python
df.head()
```

```python
# Plot the amount of Experience Years against the Salary
df.plot(kind='scatter',
        x='Experience Years',
        y='Salary',
        title='Experience Years and Salary')

# Manually plot line through middle of points
plt.plot([1.1, 10.5], [39000, 121000], color='red', linewidth=3)
```

```python
# Algorithmically draw the line
def line(x, w=0, b=0):
    return x * w + b
```

```python
x = np.linspace(1.1, 10.5, 100)
```

```python
x
```

```python
yhat = line(x, w=0, b=0)
```

```python
yhat
```

```python
df.plot(kind='scatter',
        x='Experience Years',
        y='Salary',
        title='Experience and Salary')
plt.plot(x, yhat, color='red', linewidth=3)
```

```python
def mean_squared_error(y_true, y_pred):
    s = (y_true - y_pred)**2
    return s.mean()
```

```python
X = df[['Experience Years']].values
y_true = df['Salary'].values
```

```python
X
```

```python
y_true
```

```python
y_pred = line(X)
```

```python
y_pred
```

```python
mean_squared_error(y_true, y_pred.ravel())
```

```python
plt.figure(figsize=(10, 5))

# we are going to draw 2 plots in the same figure
# first plot, data and a few lines
ax1 = plt.subplot(121)
df.plot(kind='scatter',
        x='Experience Years',
        y='Salary',
        title='Experience Years and Salary', ax=ax1)

# let's explore the cost function for a few values of b between -100 and +150
bbs = np.array([-100, -50, 0, 50, 100, 150])
mses = []  # we will append the values of the cost here, for each line
for b in bbs:
    y_pred = line(X, w=2, b=b)
    mse = mean_squared_error(y_true, y_pred)
    mses.append(mse)
    plt.plot(X, y_pred)

# second plot: Cost function
ax2 = plt.subplot(122)
plt.plot(bbs, mses, 'o-')
plt.title('Cost as a function of b')
plt.xlabel('b');
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam, SGD
```

```python
model = Sequential()
```

```python
model.add(Dense(1, input_shape=(1,)))
```

```python
model.summary()
```

```python
model.compile(Adam(learning_rate=5000), 'mean_squared_error')
```

```python
model.fit(X, y_true, epochs=40)
```

```python
y_pred = model.predict(X)
```

```python
y_pred
```

```python
df.plot(kind='scatter',
        x='Experience Years',
        y='Salary',
        title='Experience Years and Salary')
plt.plot(X, y_pred, color='red')
```
